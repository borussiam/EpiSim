```python
import epidemicsim_functions as f
from epidemics_colormap import *
import matplotlib.pyplot as plt

def elapse_ndays (rows, columns, days, alpha, beta):
    grid = f.create_grid (rows, columns, boundary = True)
    R_mid, C_mid = (rows + 1) // 2, (columns + 1) // 2
    grid[R_mid][C_mid] = 1 #    
    
    for i in range(days + 1):
        print("Day" + " "+ str(i))
        print(grid)
        infected_list = f.infected(grid)
        recovered_list = f.recovered(grid)
        print("Infected: " + str(len(infected_list)))
        print("Recovered: " + str(len(recovered_list)) + '\n')
        f.recover_all (grid, beta)
        f.infect_all (grid, alpha)
        
def draw_ndays (rows, columns, days, alpha, beta):
    grid = f.create_grid (rows, columns, boundary = True)
    R_mid, C_mid = (rows + 1) // 2, (columns + 1) // 2
    grid[R_mid][C_mid] = 1 #    
    
    for i in range(days + 1):
        plt.subplot()
        plt.imshow(grid, cmap = SIRcmap(), interpolation = 'nearest')
        plt.title("Day" + " " + str(i), fontsize = 12)
        plt.show()
        infected_list = f.infected(grid)
        recovered_list = f.recovered(grid)
        print("Infected: " + str(len(infected_list)))
        print("Recovered: " + str(len(recovered_list)) + '\n')
        f.recover_all (grid, beta)
        f.infect_all (grid, alpha)
        
draw_ndays(10, 10, 10, 0.3, 0.1)
```
